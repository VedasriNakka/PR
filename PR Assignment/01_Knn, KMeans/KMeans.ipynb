```python
import random
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

# K of the K-means
for n in range(5,15,3):
    print(n)
    NUMBEROFCLUSTERS=n


    def display_digit(digit, numberCluster):
        image = digit
        plt.figure()
        fig = plt.imshow(image.reshape(28,28))
        fig.set_cmap('gray_r')
        fig.axes.get_xaxis().set_visible(False)
        fig.axes.get_yaxis().set_visible(False)
        plt.suptitle('Centroid of the cluster nÂ° '+ str(numberCluster), fontsize=16)



    def initKRandomCentroids(k, trainingSet):
        randomDigit=[]
        for i in range(0,k):
            centroid = random.randint(0, len(trainingSet))
            randomDigit.append(centroid)
        return randomDigit

    def newCentroid(dictionary, arrayOfCentroids):
        i=0
        for indexCentroid, digits in dictionary.items():
            newCentroid = np.mean(digits, axis=0)
            arrayOfCentroids[i]= newCentroid
            dictionary[indexCentroid]=[]
            i+=1


    def kmeans(initCentroids, trainSet):
        centroidsDigit=[]
        clusters = {}
        z=0
        NUMBERITERATION=5
        for i in range(0, len(initCentroids)):
            centroidsDigit.append(trainSet[initCentroids[i]])
            clusters[i]=[]
        while z<NUMBERITERATION:
            z+=1
            for digit in trainSet:
                distRef = float('inf')
                indexRef = float('inf')
                for indexCentroid, centroid in enumerate(centroidsDigit):
                    #euclidian distance
                    computedDistance = np.linalg.norm(digit - centroid)
                    if computedDistance < distRef:
                        distRef = computedDistance
                        indexRef = indexCentroid
                clusters[indexRef].append(digit)
            if(z!=(NUMBERITERATION)):
                newCentroid(clusters, centroidsDigit)

        for index, digit in enumerate(centroidsDigit):
                display_digit(digit, index)

        return centroidsDigit, clusters


    #trainSet = np.genfromtxt("trainShort.csv", delimiter=",")
    trainSet = np.genfromtxt("C:\\Users\\Veda\\Desktop\\UniNe\\PR\\Excercise\\dataset\\TrainD.csv", delimiter=",")
    #keep all rows but skip column 1
    trainSet = trainSet[:, 1:]
    initialCentroids = initKRandomCentroids(NUMBEROFCLUSTERS, trainSet)
    allCentroids, dictionaryOfDigits= kmeans(initialCentroids, trainSet)
print("-------------")

```

```python
def dunnIndex(dictionary):
    # computation of the distance of the two closest clusters
    minDistance= float('inf')
    for i in range(0, len(dictionary)):
        for j in range(i+1, len(dictionary)):
            for k in range(0,len(dictionary[i])):
                for z in range(k+1, len(dictionary[j])):
                    distance= np.linalg.norm(dictionary[i][k]- dictionary[j][z])
                    if distance<minDistance:
                        minDistance=distance
                        
    # computation of the max diameter of all clusters
    maximumDistanceOfCluster= -float('inf')
    for index, cluster in dictionary.items():
        for i in range(0, len(cluster)):
            for j in range(i+1, len(cluster)):
                distance = np.linalg.norm(cluster[i] - cluster[j])
                if distance>maximumDistanceOfCluster:
                    maximumDistanceOfCluster = distance

    return minDistance/maximumDistanceOfCluster

dunnIndex = dunnIndex(dictionaryOfDigits)
print("The dunn index is:", dunnIndex)
```

```python
def cIndex(dictionary):
    lmbda = 0
    alpha = 0
    allDistances = []

    for i in range(0, len(dictionary)):
        for j in range(0, len(dictionary)):
            if i==j:
                for k in range(0,len(dictionary[i])):
                    for z in range(k+1, len(dictionary[j])):
                        distance= np.linalg.norm(dictionary[i][k]- dictionary[j][z])
                        allDistances.append(distance)
                        alpha+=1
                        lmbda+=distance
                    
    for i in range(0, len(dictionary)):
        for j in range(i+1, len(dictionary)):
            for k in range(0,len(dictionary[i])):
                for z in range(k+1, len(dictionary[j])):
                    distance= np.linalg.norm(dictionary[i][k]- dictionary[j][z])
                    allDistances.append(distance)

    allDistances.sort()
    mini = sum(allDistances[:alpha])
    allDistances.sort(reverse=True)
    maxi= sum(allDistances[:alpha])
    return (lmbda-mini)/(maxi-mini)

cIndex = cIndex(dictionaryOfDigits)
print("The C-index is:", cIndex)
```
