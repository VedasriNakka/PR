```python
import numpy as np
from csv import reader
 
# Load a test CSV file
def load_csv(filename):
    dataset = list()
    with open(filename, 'r') as file:
        csv_reader = reader(file)
        for row in csv_reader:
            if not row:
                continue
            dataset.append(row)
    return dataset
 
# Load dataset

#filename_test = 'C:\\Users\\Veda\\Desktop\\UniNe\\PR\\Excercise\\Done-Ex\\pattern-recognition-assignment\\01a_KNN\\test.csv'
filename_test = '..01a_KNN/test.csv'
test_dataset = load_csv(filename_test)
test_dataset = np.array((test_dataset), dtype = np.float32)
print(f"Loaded test data file {filename_test} with {len(test_dataset)} rows and {len(test_dataset[0])} columns")

#COnvert lable and formet of test
test_label = test_dataset[:, 0]
test_label = test_label.astype('int')
test_feature = test_dataset[:,1:]

# Load dataset
filename_train = 'C:\\Users\\Veda\\Desktop\\UniNe\\PR\\Excercise\\Done-Ex\\pattern-recognition-assignment\\01a_KNN\\train.csv'
train_dataset = load_csv(filename_train)
train_dataset = np.array((train_dataset), dtype = np.float32)
print(f"Loaded train data file {filename_train} with {len(train_dataset)} rows and {len(train_dataset[0])} columns")

#Convert lable and formet of train
train_label = train_dataset[:,0]
train_feature = train_dataset[:,1:]
print("Length of test formet, train formet ", len(test_feature), len(train_feature))

#L2
def euclidean(test_feature, train_feature):
    dist = np.sqrt(np.sum((test_feature - train_feature) ** 2))
    return dist 

#L1
def manhattan(test_feature, train_feature):
    dist = np.sum(np.abs(test_feature - train_feature))
    return dist 

#Cosine
def cosine(test_feature, train_feature):
    d1 = np.sum(test_feature * train_feature)
    d2 = np.sqrt(np.sum((test_feature) ** 2))
    d3 = np.sqrt(np.sum((train_feature) ** 2))
    dist = d1 / (d2 * d3) 
    return 1 - dist

def knn(k, dist_fn):
    arr_final = []
    for i in range(len(test_feature)):
        distance = []
        #print("test labels", test_label[i])
        for j in range(len(train_feature)):
            dist = dist_fn(test_feature[i],train_feature[j])
            distance.append(dist)
        #print("distance between test feature and train feature's are:{}".format(distance))
        sort_dist = np.argsort(distance)
        select = sort_dist[:k]
        #print("{}positions of Nearest k neighbours".format(select))
        top_values = train_label[select].astype(int)
        #print("{} position values of Nearest k neighbours".format(top_values))

        num_of_counts = np.bincount(top_values)
        max_label = np.argmax(num_of_counts)
        #print("{} repets most near of the K ".format(max_label))
        arr_final.append(max_label)
        #exit()


    count = 0
    for i in range(len(test_label)):
        if arr_final[i] == test_label[i]:
            #print(arr_final[i],test_label[i])
            count += 1
    mean_value = (count / len(test_label))
    print("{}-NN, Dist_fn: {}, Accuracy: {:.2f}".format(k, dist_fn.__name__, 100*mean_value))


for k in range(1,15,2):
    for dist_fn in [euclidean, manhattan, cosine]:
        knn(k, dist_fn)

# mean = np.mean(test_label == arr_final)
# print(mean)


```
